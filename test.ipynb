```python
import numpy as np

# 벡터 w와 행렬 X 정의
w = np.array([1 + 2j, 3 + 4j])  # 복소수 벡터
X = np.array([[5, 6], [7, 8]])  # 2x2 행렬

# w의 복소수 전치(conjugate transpose)와 X의 행렬 곱
X_weighted = w.conj().T @ X

# 결과 출력
print("w.conj().T:", w.conj().T)
print("X_weighted:", X_weighted)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 파라미터 설정
Nr = 4  # 안테나 수 (4개 안테나)
d = 0.5  # 안테나 간격 (0.5 파장)
N = 10000  # 신호의 샘플 수
t = np.linspace(0, 1, N)  # 1초 동안 10000개의 시간 샘플

# 가상의 신호 생성 (30도 방향에서 오는 신호)
theta_true = np.radians(30)  # 30도 방향으로 신호가 온다고 가정
signal = np.exp(1j * 2 * np.pi * 1 * t)  # 1Hz의 신호

# 수신된 신호 생성 (각각 다른 각도에서 오는 신호를 합성)
X = np.vstack([signal, signal, signal, signal])  # 4개의 안테나에서 동일한 신호 수신

# 각도를 바꿔가며 신호가 가장 강한 방향을 찾기
theta_scan = np.linspace(-np.pi, np.pi, 1000)  # -180°에서 180°까지 1000개의 각도

results = []  # 각도별 신호 강도를 저장할 리스트

# 각도에 대해 빔포밍 수행
for theta_i in theta_scan:
    # 각도에 맞는 가중치 계산 (신호가 특정 방향에서 오는 것처럼 가중치를 적용)
    w = np.exp(-2j * np.pi * d * np.arange(Nr) * np.sin(theta_i))  # 가중치 계산
    X_weighted = w.conj().T @ X  # 가중치를 적용한 신호 합성
    results.append(10 * np.log10(np.var(X_weighted)))  # 신호 강도 (dB로 변환)

# 결과 정규화
results -= np.max(results)  # 최대값을 0으로 정규화

# 가장 강한 신호가 오는 각도 출력
detected_angle = theta_scan[np.argmax(results)] * 180 / np.pi  # 최대값을 주는 각도 (도 단위로 변환)
print(f"Detected Angle: {detected_angle}°")  # 결과 출력

# 결과 플로팅
plt.plot(theta_scan * 180 / np.pi, results)  # 각도를 도 단위로 변환하여 플로팅
plt.xlabel("Theta [Degrees]")
plt.ylabel("Signal Strength (dB)")
plt.title("Beamforming: Finding the Strongest Signal Direction")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 파라미터 설정
Nr = 4  # 안테나 수 (4개 안테나)
d = 0.5  # 안테나 간격 (0.5 파장)
N = 1000  # 신호의 샘플 수 (1000개의 시간 샘플)
t = np.linspace(0, 1, N)  # 1초 동안 1000개의 시간 샘플

# 가상의 신호 생성 (30도 방향에서 오는 신호)
theta_true = np.radians(30)  # 30도 방향으로 신호가 온다고 가정
signal = np.exp(1j * 2 * np.pi * 1 * t)  # 1Hz의 신호

# 수신된 신호 생성 (각각 다른 각도에서 오는 신호를 합성)
X = np.vstack([signal, signal, signal, signal])  # 4개의 안테나에서 동일한 신호 수신

# 수신된 신호 플로팅
plt.figure(figsize=(10, 6))
plt.plot(t, np.real(X[0]), label="Antenna 1 (Real part)")
plt.plot(t, np.imag(X[0]), label="Antenna 1 (Imaginary part)")
plt.title("Received Signal at Antenna 1 (Real and Imaginary Parts)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()

# 각도를 바꿔가며 신호가 가장 강한 방향을 찾기
theta_scan = np.linspace(-np.pi, np.pi, 1000)  # -180°에서 180°까지 1000개의 각도

results = []  # 각도별 신호 강도를 저장할 리스트

# 각도에 대해 빔포밍 수행
for theta_i in theta_scan:
    # 각도에 맞는 가중치 계산 (신호가 특정 방향에서 오는 것처럼 가중치를 적용)
    w = np.exp(-2j * np.pi * d * np.arange(Nr) * np.sin(theta_i))  # 가중치 계산
    w = w[:, np.newaxis]  # w의 차원을 (4, 1)로 바꾸어 X와 맞추기
    
    X_weighted = np.sum(w * X, axis=0)  # 가중치를 적용한 신호 합성 (여기서는 벡터로 합성)
    results.append(10 * np.log10(np.var(X_weighted)))  # 신호 강도 (dB로 변환)

# 결과 정규화
results -= np.max(results)  # 최대값을 0으로 정규화

# 결과 플로팅
plt.figure(figsize=(10, 6))
plt.plot(theta_scan * 180 / np.pi, results)  # 각도를 도 단위로 변환하여 플로팅
plt.xlabel("Theta [Degrees]")
plt.ylabel("Signal Strength (dB)")
plt.title("Beamforming: Finding the Strongest Signal Direction")
plt.grid()
plt.show()

# 가장 강한 신호가 오는 각도 출력
detected_angle = theta_scan[np.argmax(results)] * 180 / np.pi  # 최대값을 주는 각도 (도 단위로 변환)
print(f"Detected Angle: {detected_angle}°")  # 결과 출력
```

```python
import numpy as np
import matplotlib.pyplot as plt

sample_rate = 1e6
N = 10000  # number of samples to simulate

# Create a tone to act as the transmitter signal
t = np.arange(N) / sample_rate  # time vector
f_tone = 0.02e6
tx = np.exp(2j * np.pi * f_tone * t)  # Transmitter signal

d = 0.5  # half wavelength spacing
Nr = 3  # Number of receiver elements
theta_degrees = 20  # Direction of arrival (feel free to change this, it's arbitrary)
theta = theta_degrees / 180 * np.pi  # Convert to radians

# Steering vector (3x1)
s = np.exp(-2j * np.pi * d * np.arange(Nr) * np.sin(theta))

s = s.reshape(-1, 1)  # Make s a column vector (3x1)
tx = tx.reshape(1, -1)  # Make tx a row vector (1x10000)

# Simulate the received signal X through matrix multiplication
X = s @ tx  # X will have dimensions (3, 10000)
print(X.shape)  # Should print (3, 10000)

# Add noise to the received signal
n = np.random.randn(Nr, N) + 1j * np.random.randn(Nr, N)  # Complex noise
X = X + 0.2 * n  # Add noise to the received signal

# Plot the real part of the received signal for the first 200 samples
plt.plot(np.asarray(X[0, :]).squeeze().real[0:200])
plt.plot(np.asarray(X[1, :]).squeeze().real[0:200])
plt.plot(np.asarray(X[2, :]).squeeze().real[0:200])
plt.show()

# Conventional delay-and-sum beamformer weights
w = np.exp(-2j * np.pi * d * np.arange(Nr) * np.sin(theta))

# Apply the weights to the received signal (beamforming)
X_weighted = w.conj().T @ X  # Perform beamforming (1x10000)
print(X_weighted.shape)  # Should print (1, 10000)

# Scan over different angles and compute the beamforming output
theta_scan = np.linspace(-np.pi, np.pi, 1000)  # 1000 different thetas between -180 and +180 degrees
results = []

for theta_i in theta_scan:
    w = np.exp(-2j * np.pi * d * np.arange(Nr) * np.sin(theta_i))  # Compute weights for the current angle
    X_weighted = w.conj().T @ X  # Apply the weights to the received signal
    results.append(10 * np.log10(np.var(X_weighted)))  # Power in dB

results -= np.max(results)  # Normalize the results (optional)

# Print the angle that gave us the max value
print(theta_scan[np.argmax(results)] * 180 / np.pi)  # Angle in degrees

# Plot the results
plt.plot(theta_scan * 180 / np.pi, results)  # Plot angle in degrees
plt.xlabel("Theta [Degrees]")
plt.ylabel("DOA Metric")
plt.grid()
plt.show()
```
